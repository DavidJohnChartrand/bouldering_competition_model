```python
import pandas as pd
```

```python
path = "resources/boulder_results.csv"
df = pd.read_csv(path)
df.head()
```

```python
df.columns
```

```python
df['Q_Tops']=df["Qualification"].str.slice(start=0,stop=1)
df['Q_Zones']=df["Qualification"].str.slice(start=2,stop=3)
df['Q_Attempts']=df["Qualification"].str.slice(start=4)
df['Q_Top_attempts']=df['Q_Attempts'].str.slice(start=0,stop=1)
df['Q_Zone_attempts']=df['Q_Attempts'].str.slice(start=1)
df['S_Tops']=df["Semifinal"].str.slice(start=0,stop=1)
df['S_Zones']=df["Semifinal"].str.slice(start=2,stop=3)
df['S_Attempts']=df["Semifinal"].str.slice(start=4)
df['S_Top_attempts']=df['S_Attempts'].str.slice(start=0,stop=1)
df['S_Zone_attempts']=df['S_Attempts'].str.slice(start=1)
df['F_Tops']=df["Final"].str.slice(start=0,stop=1)
df['F_Zones']=df["Final"].str.slice(start=2,stop=3)
df['F_Attempts']=df["Final"].str.slice(start=4)
df['F_Top_attempts']=df['F_Attempts'].str.slice(start=0,stop=1)
df['F_Zone_attempts']=df['F_Attempts'].str.slice(start=1)
df = df.drop(columns=['Q_Attempts','S_Attempts','F_Attempts'])
df
```

```python
df.to_csv("resources/cleaned_scores.csv")
```
